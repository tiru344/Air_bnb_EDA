```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=df.dropna()
```

```python
df.isnull().sum()
```

```python
df[df.duplicated()]
```

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
df.dtypes
```

```python
df['id']=df['id'].astype(object)

df['host_id']=df['host_id'].astype(object)
df.dtypes
```

```python
df=df[df['price']<1500]
sns.boxplot(data=df,x='price')
```

```python
plt.figure(figsize=(4,5))
sns.histplot(data=df,x='price',bins=100)
plt.title("Price History")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.histplot(data=df,x='availability_365')
plt.title("Price History")
plt.ylabel("Frequency")
plt.show()
```

```python
df.groupby('neighbourhood_group')['price'].mean()
```

```python
df.columns
```

```python
sns.barplot(data=df,x='neighbourhood_group',y='price',hue='room_type')
plt.show()
```

```python
plt.Figure(figsize=(4,6))
sns.scatterplot(data=df,x='number_of_reviews',y='price',hue='neighbourhood_group')
plt.show()
```

```python
sns.pairplot(data=df,vars=['price','number_of_reviews','availability_365', 'minimum_nights'],hue='room_type')
plt.show()
```
